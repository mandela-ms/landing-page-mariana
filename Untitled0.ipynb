```python
# Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ============================
# 1 - Série: número de alunos por titulação
Titulação = pd.Series(
    ['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'],
    index=[20, 15, 85, 145, 320]
)
print("\n1) Série - Número de alunos por titulação:")
print(Titulação)

# ============================
# 2 - DataFrame: número de alunos por titulação
df_titulacao = pd.DataFrame({
    'Titulação': ['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'],
    'Quantidade': [20, 15, 85, 145, 320]
})
print("\n2) DataFrame - Número de alunos por titulação:")
print(df_titulacao)

# ============================
# 3 - Tabela de calorias e gordura
df_alimentos = pd.DataFrame({
    'calorias': [200, 350, 550],
    'gordura (%)': [0, 15, 35]
}, index=['banana', 'macarrão', 'cachorro quente'])
print("\n3) DataFrame - Calorias e gordura:")
print(df_alimentos)

# ============================
# 4 - Matriz com 50 elementos e 3 dimensões
v = np.array(range(50)).reshape(2, 5, 5)
print("\n4) Matriz 3D:")
print('Shape = ', v.shape)
print('Número de dimensões = ', v.ndim)
print('Número de elementos = ', v.size)
print('Tensor v =\n', v)

# ============================
# 5 - Gráfico de pizza
fatias = [8, 12, 10]
atividades = ['Lazer', 'Estudo', 'Viagem']
cores = ['r', 'm', 'y']
plt.figure()
plt.pie(fatias, labels=atividades, colors=cores, startangle=90, shadow=True, explode=(0, 0.1, 0))
plt.title("5) Gráfico de Pizza")
plt.show()

# ============================
# 6 - Gráfico de linha
x = [1, 3, 5]
y = [1, 2, 5]
plt.figure()
plt.plot(x, y, label='Uma legenda')
plt.title('6) Exemplo utilizando Plot')
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')
plt.legend()
plt.show()

# ============================
# 7 - Gráfico de dispersão com formato de estrela
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]
plt.figure()
plt.scatter(x, y, label='Pontos', color='b', marker='*', s=100)
plt.legend()
plt.title("7) Gráfico de Dispersão - Estrelas")
plt.show()
```
